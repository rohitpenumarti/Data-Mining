```python
from sklearn.preprocessing import LabelEncoder

all_city_states = business_df['city']+'-'+business_df['state']
unique_city_states = list(set(all_city_states.values))

unique_names = list(set(business_df['name'].values))

le_cs = LabelEncoder()
le_cs.fit(unique_city_states)
print(le_cs.classes_)

le_n = LabelEncoder()
le_n.fit(unique_names)
print(le_n.classes_)
```

```python
new_train_df['city_state_le'] = le_cs.transform(new_train_df['city_state'])
new_train_df['name_le'] = le_n.transform(new_train_df['name'])
new_train_df.head(5)
```

```python
X_train = new_train_df[['rounded_business_rating', 'latitude', 'longitude', 'average_user_rating', 'num_user_reviews', 'avg_user_compliments', \
    'bike_parking', 'business_accepts_credit_cards', 'good_for_kids', 'has_TV', 'outdoor_seating', 'restaurants_delivery', \
        'restaurants_good_for_groups', 'restaurants_reservations', 'restaurants_take_out', \
            'price_range', 'city_state_le', 'name_le', 'reviews_per_year', 'garage', 'street', 'validated', 'lot', 'valet',]]
y_train = new_train_df['rating']

xgb = XGBRegressor().fit(X_train, y_train)

y_pred_train = list(xgb.predict(X_train))
train_rmse = mean_squared_error(list(y_train), y_pred_train, squared=False)

X_test = final_test_df[['rounded_business_rating', 'latitude', 'longitude', 'average_user_rating', 'num_user_reviews', 'avg_user_compliments', \
    'bike_parking', 'business_accepts_credit_cards', 'good_for_kids', 'has_TV', 'outdoor_seating', 'restaurants_delivery', \
        'restaurants_good_for_groups', 'restaurants_reservations', 'restaurants_take_out', \
            'price_range', 'city_state_le', 'name_le', 'reviews_per_year', 'garage', 'street', 'validated', 'lot', 'valet',]]
y_test = final_test_df[' rating']
y_pred_test = list(xgb.predict(X_test))
test_rmse = mean_squared_error(list(y_test), y_pred_test, squared=False)

print(f'Train RMSE: {train_rmse}; Test RMSE: {test_rmse}')
```

```python
from sklearn.feature_selection import RFE

n = len(list(X_train.keys()))-1
model_performances_with_features = {}
while n > 0:
    test_xgb = XGBRegressor()
    rfe = RFE(test_xgb, n_features_to_select=n)

    rfe.fit(X_train, y_train)
    features_supp = rfe.support_
    features_out = [list(X_train.keys())[i] for i, supp in enumerate(features_supp) if supp == True]

    test_xgb.fit(X_train[features_out], y_train)
    y_pred_train_rfe = list(test_xgb.predict(X_train[features_out]))
    train_rmse_rfe = mean_squared_error(list(y_train), y_pred_train_rfe, squared=False)

    y_pred_test_rfe = list(test_xgb.predict(X_test[features_out]))
    test_rmse_rfe = mean_squared_error(list(y_test), y_pred_test_rfe, squared=False)
    print(f'Train RMSE: {train_rmse_rfe}; Test RMSE: {test_rmse_rfe}')
    n -= 1
    model_performances_with_features[(train_rmse_rfe, test_rmse_rfe)] = features_out

best_model = min(model_performances_with_features.items(), key=lambda x: x[0][1])
print(f'Best Model:\nTrain RMSE: {best_model[0][0]}; Test RMSE: {best_model[0][1]}\nBest Features: {best_model[1]}')
```
